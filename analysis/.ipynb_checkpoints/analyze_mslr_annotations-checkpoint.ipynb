```python
import os
import csv
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from collections import defaultdict, Counter
from sklearn.metrics import cohen_kappa_score
from rouge import Rouge 

ANNOTATION_DATA_DIR = os.path.expanduser('../data/annotations/')
ANN_FILES = {
    1: os.path.join(ANNOTATION_DATA_DIR, 'ann1_incomplete.tsv'),
    2: os.path.join(ANNOTATION_DATA_DIR, 'ann2_incomplete.tsv')
}

KEY_MAP = {
    "Is the generated summary fluent?": "Fluency",
    "Is the *population* in the generated summary the same as the population in the target summary?": "Population",
    "Is the *intervention* in the generated summary the same as the intervention in the target summary?": "Intervention",
    "Is the *outcome* in the generated summary the same as the outcome in the target summary?": "Outcome",
    "Comments about PIO agreement (optional)": "PIO_comment",
    "What is the effect direction in the *target* summary?": "Target Effect Direction",
    "What is the effect direction in the *generated* summary?": "Generated Effect Direction",
    "Comments about effect directions (optional)": "ED_comment",
    "What is the strength of the claim made in the *target* summary?": "Target Strength",
    "What is the strength of the claim made in the *generated* summary?": "Generated Strength",
    "Comments about strength of claim (optional)": "Strength_comment"    
}

EXP_MAPPINGS = {
    '01G4NE2DDS5G6Q047M97PX7SGV': 'AI2/BART/train Cochrane/decode Cochrane',
    '01G8WPZRN2E3EHA2WENHVNCH8M': 'ittc1',
    '01G9JE4STYHQ2136MCATAQ85CE': 'bart-large-finetuned',
    '01G9RKHTAQVPR038VTDCJB6Z8F': 'ittc2',
    '01GA1HEQEJHQHEAQD8YX8FWF5T': 'SciSpace',
    '01GCRZERDX9XKMDWQ5GDSPNXTA': 'led-base-16384-cochrane'
}

FLUENCY_MAPPINGS = {
    '2: Yes--there are no errors that impact comprehension of the summary': 2,
    '1: Somewhat--there are some grammatical or lexical errors but I can understand the meaning': 1,
    '0: No--there are major grammatical or lexical errors that impact comprehension': 0
}

POPULATION_MAPPINGS = {
    '2: Yes': 2,
    '1: Partially': 1,
    '0: No': -1,
    'Other / uncertain (please comment)': np.nan,
    'N/A: No population in generated summary': 0,
}

INTERVENTION_MAPPINGS = {
    '2: Yes': 2,
    '1: Partially': 1,
    '0: No': -1,
    'Other / uncertain (please comment)': np.nan,
    'N/A: No intervention in generated summary': 0,
}

OUTCOME_MAPPINGS = {
    '2: Yes': 2,
    '1: Partially': 1,
    '0: No': -1,
    'Other / uncertain (please comment)': np.nan,
    'N/A: No outcome in generated summary': 0,
}

TARGET_DIRECTION_MAPPINGS = {
    '(-1): Negative effect': -1,
    '0: No effect': 0,
    '(+1): Positive effect': 1,
    'N/A: no effect direction is specified in the target summary': 0,
    'Other / uncertain (please comment)': np.nan
}

GENERATED_DIRECTION_MAPPINGS = {
    '(-1): Negative effect': -1,
    '0: No effect': 0,
    '(+1): Positive effect': 1,
    'N/A: no effect direction is specified in the generated summary': 0,
    'N/A: no effect direction is specified in the target summary': 0,
    'Other / uncertain (please comment)': np.nan
}

TARGET_STRENGTH_MAPPINGS = {
    '0: Not enough evidence (there is insufficient evidence to draw a conclusion)': 0,
    '1: Weak claim': 1,
    '2: Moderate claim': 2,
    '3: Strong claim': 3,
    'N/A: No claim (there is no claim in the target summary)': -1,    
    'Other / uncertain (please comment)': np.nan,
    '': np.nan
}

GENERATED_STRENGTH_MAPPINGS = {
    '0: Not enough evidence (there is insufficient evidence to draw a conclusion)': 0,
    '1: Weak claim': 1,
    '2: Moderate claim': 2,
    '3: Strong claim': 3,
    'Other / uncertain (please comment)': np.nan
}
```

```python
# load annotation data
annotations = []
iaa_data = defaultdict(dict)

for ind, fname in ANN_FILES.items():
    with open(fname, 'r') as f:
        reader = csv.DictReader(f, delimiter='\t')
        for row in reader:
            row['Annotator'] = ind
            row['UID'] = f"{row['ExpID']}_{row['ReviewID']}"
            annotations.append(row)
        
for row in annotations:
    anum = row['Annotator']
    if not(row["Is the generated summary fluent?"]):
        continue
    for key, value in row.items():
        if key in KEY_MAP:
            new_key = f'{KEY_MAP[key]}_{anum}'
            iaa_data[row['UID']][new_key] = value
        
df = pd.DataFrame(annotations)
df = df.rename(columns=KEY_MAP)
df = df[df['Fluency'] != '']
```

```python
# add analysis values
df['System'] = df.apply(lambda row: EXP_MAPPINGS[row.ExpID], axis=1)
df['fluency_yes'] = df.apply(lambda row: int(row['Fluency'] == '2: Yes--there are no errors that impact comprehension of the summary'), axis=1)
df['fluency_partial'] = df.apply(lambda row: int(row['Fluency'] == '1: Somewhat--there are some grammatical or lexical errors but I can understand the meaning'), axis=1)
df['population_yes'] = df.apply(lambda row: int(row['Population'] == '2: Yes'), axis=1)
df['population_partial'] = df.apply(lambda row: int(row['Population'] == '1: Partially'), axis=1)
df['intervention_yes'] = df.apply(lambda row: int(row['Intervention'] == '2: Yes'), axis=1)
df['intervention_partial'] = df.apply(lambda row: int(row['Intervention'] == '1: Partially'), axis=1)
df['outcome_yes'] = df.apply(lambda row: int(row['Outcome'] == '2: Yes'), axis=1)
df['outcome_partial'] = df.apply(lambda row: int(row['Outcome'] == '1: Partially'), axis=1)
df['PIO_yes'] = df.apply(lambda row: int(row.population_yes and row.intervention_yes and row.outcome_yes), axis=1)
df['PIO_partial'] = df.apply(lambda row: int((row.population_yes or row.population_partial) and \
                                 (row.intervention_yes or row.intervention_partial) and \
                                 (row.outcome_yes or row.outcome_partial)), 
                            axis=1)
df['effect_direction_equal'] = df.apply(lambda row: int(row['Target Effect Direction'] == row['Generated Effect Direction']), axis=1)
df['strength_equal'] = df.apply(lambda row: int(row['Target Strength'] == row['Generated Strength']), axis=1)
df['all_good'] = df.apply(lambda row: int(row.fluency_yes and row.PIO_yes and row.effect_direction_equal and row.strength_equal), axis=1)
df['pio_okay'] = df.apply(lambda row: int((row.fluency_yes or row.fluency_partial) and \
                          (row.PIO_yes or row.PIO_partial)), axis=1)
df['all_okay'] = df.apply(lambda row: int((row.fluency_yes or row.fluency_partial) and \
                          (row.PIO_yes or row.PIO_partial) and row.effect_direction_equal), axis=1)
df['num_good'] = df.apply(lambda row: int(row.fluency_yes) + int(row.PIO_yes) + int(row.effect_direction_equal) + int(row.strength_equal), axis=1)
df['num_okay'] = df.apply(lambda row: int(row.fluency_yes or row.fluency_partial) + \
                          int(row.PIO_yes or row.PIO_partial) + int(row.effect_direction_equal) + int(row.strength_equal), axis=1)

df['fluency_value'] = df.apply(lambda row: FLUENCY_MAPPINGS.get(row['Fluency'], np.nan), axis=1)
df['population_value'] = df.apply(lambda row: POPULATION_MAPPINGS.get(row['Population'], np.nan), axis=1)
df['intervention_value'] = df.apply(lambda row: INTERVENTION_MAPPINGS.get(row['Intervention'], np.nan), axis=1)
df['outcome_value'] = df.apply(lambda row: OUTCOME_MAPPINGS.get(row['Outcome'], np.nan), axis=1)
df['pio_mean'] = df.apply(lambda row: np.mean([row['population_value'], row['intervention_value'], row['outcome_value']]), axis=1)
df['target_direction_value'] = df.apply(lambda row: TARGET_DIRECTION_MAPPINGS.get(row['Target Effect Direction'], np.nan), axis=1)
df['generated_direction_value'] = df.apply(lambda row: GENERATED_DIRECTION_MAPPINGS.get(row['Generated Effect Direction'], np.nan), axis=1)
df['direction_diff'] = df.apply(lambda row: np.abs(row['target_direction_value'] - row['generated_direction_value']), axis=1)
df['target_strength_value'] = df.apply(lambda row: TARGET_STRENGTH_MAPPINGS.get(row['Target Strength'], np.nan), axis=1)
df['generated_strength_value'] = df.apply(lambda row: GENERATED_STRENGTH_MAPPINGS.get(row['Generated Strength'], np.nan), axis=1)
df['strength_diff'] = df.apply(lambda row: np.abs(row['target_strength_value'] - row['generated_strength_value']), axis=1)

```

```python
# add automated metrics
rouge = Rouge()
df['rouge'] = df.apply(lambda row: rouge.get_scores(row['Generated Summary'], row['Target Summary']), axis=1)
df['rouge1'] = df.apply(lambda row: row['rouge'][0]['rouge-1']['f'], axis=1)
df['rouge2'] = df.apply(lambda row: row['rouge'][0]['rouge-2']['f'], axis=1)
df['rougeL'] = df.apply(lambda row: row['rouge'][0]['rouge-l']['f'], axis=1)
```

```python
col_to_keep = [
    'UID', 'System', 'ReviewID',
    'Target Summary', 'Generated Summary',
    'all_good', 'pio_okay', 'all_okay', 'num_good', 'num_okay', 
    'fluency_value', 
    'population_value', 'intervention_value', 'outcome_value', 'pio_mean',
    'target_direction_value', 'generated_direction_value', 'direction_diff',
    'target_strength_value', 'generated_strength_value', 'strength_diff',
    'rouge1', 'rouge2', 'rougeL'
]

df = df[col_to_keep]
display(df)
```

```python
# Inter-annotator agreement
iaa_data = {k: v for k, v in iaa_data.items() if len(v) > 11}
print('Overlapping annotations:', len(iaa_data))
print()

for k, v in iaa_data.items():
    v['UID'] = k
    
df_iaa = pd.DataFrame(iaa_data.values())

kappa_fluency = cohen_kappa_score(df_iaa['Fluency_1'], df_iaa['Fluency_2'])
kappa_population = cohen_kappa_score(df_iaa['Population_1'], df_iaa['Population_2'])
kappa_intervention = cohen_kappa_score(df_iaa['Intervention_1'], df_iaa['Intervention_2'])
kappa_outcome = cohen_kappa_score(df_iaa['Outcome_1'], df_iaa['Outcome_2'])
kappa_tdirection = cohen_kappa_score(df_iaa['Target Effect Direction_1'], df_iaa['Target Effect Direction_2'])
kappa_gdirection = cohen_kappa_score(df_iaa['Generated Effect Direction_1'], df_iaa['Generated Effect Direction_2'])
kappa_tstrength = cohen_kappa_score(df_iaa['Target Strength_1'], df_iaa['Target Strength_2'])
kappa_gstrength = cohen_kappa_score(df_iaa['Generated Strength_1'], df_iaa['Generated Strength_2'])
mean_kappa = np.mean([
    kappa_fluency,
    kappa_population,
    kappa_intervention,
    kappa_outcome,
    kappa_tdirection,
    kappa_gdirection,
    kappa_tstrength,
    kappa_gstrength
])

print('=== Inter Annotation Agreement ===')
print('Fluency: ', kappa_fluency)
print('Population: ', kappa_population)
print('Intervention: ', kappa_intervention)
print('Outcome: ', kappa_outcome)
print('Target direction: ', kappa_tdirection)
print('Generated direction: ', kappa_gdirection)
print('Target strength: ', kappa_tstrength)
print('Generated strength: ', kappa_gstrength)
print()
print('Mean kappa: ', mean_kappa)

```

```python
# Correlation between human eval facets and RougeL
xs = [
    'all_okay', 
    'num_okay', 
    'fluency_value', 
    'pio_mean',
    'direction_diff', 
    'strength_diff',
]
ys = ['rougeL']

sns.pairplot(data=df, kind="reg", y_vars=ys, x_vars=xs)
# sns.pairplot(data=df, hue="System", y_vars=ys, x_vars=xs)
plt.show()
```

```python
x_human = [
    'fluency_value', 
    'pio_mean',
    'direction_diff', 
    'strength_diff',
]
sns.pairplot(data=df, x_vars=x_human, y_vars=x_human)
# sns.pairplot(data=df, hue="System", y_vars=ys, x_vars=xs)
plt.show()
```

```python
from scipy.stats import pearsonr

print('=== Pearsons correlations ===')
y = 'rougeL'
for x in xs:
    x_col = np.array(df[x])
    y_col = np.array(df[y])
    bad = ~np.logical_or(np.isnan(x_col), np.isnan(y_col))
    x_use = np.compress(bad, x_col)
    y_use = np.compress(bad, y_col)
    corr, pval = pearsonr(x_use, y_use)
    print(f'{x}: r={corr:.3f} (p={pval:.3f})')
```

```python
print('=== Pearsons correlations ===')
for y in x_human:
    for x in x_human:
        if y == x:
            continue
        x_col = np.array(df[x])
        y_col = np.array(df[y])
        bad = ~np.logical_or(np.isnan(x_col), np.isnan(y_col))
        x_use = np.compress(bad, x_col)
        y_use = np.compress(bad, y_col)
        corr, pval = pearsonr(x_use, y_use)
        print(f'{y} v. {x}: r={corr:.3f} (p={pval:.3f})')
```
